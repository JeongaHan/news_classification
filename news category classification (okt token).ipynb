```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xval,Ytrain,Yval = train_test_split(Xpad,YoneHot,test_size=0.2,random_state=807)
```

```python
wordsize= len(token.word_index)+1
```

```python
wordsize
```

```python
model = Sequential()
model.add(Embedding(wordsize,300,input_length=35))
model.add(LSTM(128,activation='tanh',return_sequences=True))
model.add(LSTM(64,activation='tanh',return_sequences=True))
model.add(LSTM(32))
model.add(Flatten())
model.add(Dense(6,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(Xtrain,Ytrain,batch_size=100,epochs=30,validation_data=(Xval,Yval))
```

```python
test_news['predicted'] = 0
for i in range(len(test_news)):
    sentence = test_news.title[i]
    hangul = re.compile('[^0-9a-zA-Zㄱ-힗]+')
    sentence = hangul.sub(" ",sentence)
    sentence = okt.morphs(sentence)
    result = []
    for j in range(len(sentence)):
        if sentence[j] not in stopwords:
            result.append(sentence[j])
    sentence = result
    result = ""
    for k in range(len(sentence)):
        result = result + " " + sentence[k]
    sentence = result
    sentence = token.texts_to_sequences([sentence])
    sent_pad = pad_sequences(sentence,35)
    test_news['predicted'][i] = np.argmax(model.predict(sent_pad))
```
