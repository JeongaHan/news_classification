```python
import pandas as pd
import ssl
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = "https://www.ranks.nl/stopwords/korean"
resp = requests.get(url, verify=False)
soup = BeautifulSoup(resp.text, 'lxml')
rows = soup.select('td')
print('rows :', rows)
print('str(rows[0]) :', str(rows[0]))
print('rows[0].text :', rows[0].text)
```

```python
bad_word_list = []
bad_word = ''
for row in rows:
    try:
        bad_word_list += (re.compile('[가-힣]+').findall(str(row)))
        #bad_word += re.compile('[^가-힣]+').sub(' ', str(row))
        print(type(bad_word))
        
    except AttributeError as err:
        pass
#bad_word = bad_word.split(' ')
print(bad_word_list)
```

```python
df = pd.DataFrame(bad_word_list)
df.columns = ['stopword']
df.to_csv('stop_word.csv', encoding='utf-8')
```
