```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('news_titles.csv')
df2 = pd.read_csv('news_titles_2.csv')
```

```python
df = pd.concat([df1,df2])
```

```python
df = df[df.title!='동영상기사']
df = df[df.title!=' 기준 ']
df = df[df.title!=' 검색상위 ']
```

```python
df.to_csv('news_titles_concat.csv',index=False)
```
